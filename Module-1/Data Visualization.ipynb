```python
#import matplotlib.pyplot
import matplotlib.pyplot as plt
import numpy as np

#draw plot as fig
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')
#text annotation to the plot where it indicate maximum value of the curv
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))
#text annotation to the plot where it indicate minimum value of the curv
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"));
```

# Adding Annotatae & legend

```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt 
#you can also right

x=np.arange(0,20)
y=x^3
z=x^4

plt.plot(x,y)
plt.plot(y,z)
plt.plot(z,x)
plt.title("graph")
plt.xlabel("Time")
plt.ylabel("Distance")


#Annotataion

plt.annotate(xy=[2,1],text="first value",rotation=2)
plt.annotate(xy=[4,6],text="second value",rotation=2)
plt.annotate(xy=[8,10],text="Third value",rotation=2)

#add legend
plt.legend(["Race 1","Race 2","race 3"],loc=2)
plt.show()
```

# Pie chart

pie chart is a circular stastical plot that can display only one series of data.

the area of the chart is the total% of the given data

```python
#import librarties

import numpy as np
import matplotlib.pyplot as plt

#Creating dataset
Branch = ['computer', 'mechnical', 'electical',
		'civil', 'chemical', 'automobile']

data = [43, 17, 30, 29, 12,20]
explode=(0.1,0,0,0.3,0,0)
# Creating chart

fig = plt.figure(figsize =(12, 7))
plt.pie(data, labels = Branch,explode=explode,autopct='%1.3f%%',shadow=True)
plt.title("Neotech Branch")


# show plot
plt.show()
```

# 3D ploting

```python
#import mplot3D tiilkit,numpy &matplotlib

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3D toolkit lib
```

```python
fig=plt.figure()
# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# defining all 3 axes
z = np.linspace(0, 3, 200)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

#3D plotting
ax.plot3D(x, y, z, 'red')
ax.set_title('3D ploting')
plt.show()
```
